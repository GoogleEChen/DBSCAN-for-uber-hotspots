```python
import time
import numpy as np
import pandas as pd
%matplotlib inline
import csv 
import matplotlib.pyplot as plt
import mplleaflet as mpl # 地圖視覺化
from sklearn.cluster import DBSCAN 
```

```python
data = []
with open('uber-raw-data-aug14 (2).csv', newline='\n') as csvfile:  
    rows = csv.DictReader(csvfile)
    for row in rows:
        if row['Date/Time'].startswith('8/15/2014'):
            data.append([ float( row['Lat'] ), float( row['Lon'] )])
            
            
            
                         
                         
        
```

```python
data = np.array(data)
data.shape
```

```python
%%time

kms_per_radian = 6371.0088
epsilon = 0.05 / kms_per_radian
min_pickups = 25
```

```python
dbscan = DBSCAN(eps = epsilon , min_samples = 25, metric = 'haversine').fit(np.radians(data))

cluster_labels = dbscan.labels_
num_clusters = len(set(cluster_labels))
```

```python
print(num_clusters)
```

```python
hot_spot = []
hot_spot_num = []

for c in range(num_clusters-1):
    if c != -1:
        tmp = data[np.where(cluster_labels == c)]
        #print(tmp)
        #print('\n')
        hot_spot.append(np.mean(tmp,axis=0))
        
        hot_spot_num.append(len(tmp))
        
hot_spot = np.array(hot_spot)
hot_spot_num = np.array(hot_spot_num)
```

```python
## make the figure
fig = plt.figure(figsize = (20,15))
ax = fig.add_subplot(111)

## make the scatter plot
for idx in range(len(hot_spot)):
    plt.scatter(hot_spot[idx][1],hot_spot[idx][0],s=hot_spot_num[idx])

## display the Leaflet
#mpl.show()     # opens in a new interactive tab
mpl.display()  # shows interactive map inline in Jupyter but cannot handle large data sets
```
